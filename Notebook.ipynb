# Plotting Correlation Between Heel Hooks, 'Leg Locks', and Injuries

### First we'll take a look at the correlation between our counts of Heel Hooks and Leg Locks with respect to the total injuries in the same year. I opted to cut off fight data prior to 1998 due to small sample sizes in those years.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('cleaned_data.csv')

col_a = data.loc[data['Year'] >= 1998, 'Heel_Hook_Count']
col_b = data.loc[data['Year'] >= 1998, 'Injury_Count']

corr = col_a.corr(col_b)

print(f"The correlation between Heel_Hook_Count and Injury_Count is: {corr}")

plt.scatter(col_a, col_b)
plt.title("Heel_Hook_Count vs. Injury_Count")
plt.xlabel("Heel_Hook_Count")
plt.ylabel("Injury_Count")
plt.plot(col_a, col_a*corr, color='red')
plt.show()
```

```python
col_a = data.loc[data['Year'] >= 1998, 'Leg_Lock_Count']
col_b = data.loc[data['Year'] >= 1998, 'Injury_Count']

corr = col_a.corr(col_b)

print(f"The correlation between Leg_Lock_Count and Injury_Count is: {corr}")

plt.scatter(col_a, col_b)
plt.title("Leg_Lock_Count vs. Injury_Count")
plt.xlabel("Leg_Lock_Count")
plt.ylabel("Injury_Count")
plt.plot(col_a, col_a*corr, color='red')
plt.show()
```
